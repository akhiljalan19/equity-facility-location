# Facility Overlay

This notebook will generate a map of Alameda/Contra Costa counties with our facilities laid over it. 

## Imports

```python
from scipy.io import loadmat
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
alameda_df = pd.read_csv("all-data/hospital-data/alameda-hospitals-only.csv")
contra_costa_df = pd.read_csv("all-data/hospital-data/cc-healthcare-dataset.csv")
```

```python
alameda_lats = [row["LATITUDE"] for _, row in alameda_df.iterrows()]
alameda_longs = [row["LONGITUDE"] for _, row in alameda_df.iterrows()]
contra_costa_lats = [row["LATITUDE"] for _, row in contra_costa_df.iterrows()]
contra_costa_longs = [row["LONGITUDE"] for _, row in contra_costa_df.iterrows()]
print(len(alameda_lats))
print(len(contra_costa_lats))
```

We will set the colors and markers of Alameda facility dots.

Red = Currently operating ER. 

Blue = Hospital with no current ER service.

```python
alameda_colors = ["r" if i >= 13 else "b" for i in range(17)]
alameda_markers = ["s" if i >= 13 else "o" for i in range(17)]
print(alameda_colors)
```

```python
fig = plt.figure(figsize=(36, 20))
plt.scatter(alameda_longs[:17], alameda_lats[:17], marker="o", color=alameda_colors, s=196)
plt.scatter(contra_costa_longs, contra_costa_lats, marker="o", color="b", s=196)
ax = plt.axes()
plt.xlabel("Longitude", fontsize=36)
plt.ylabel("Latitude", fontsize=36)
ax.ticklabel_format(style='plain', useOffset=False)

for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(36)
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(36)

# show map in background
img = plt.imread("all-data/gmaps-images/alameda_cc_map_resized.png")
ax.imshow(img, alpha=.5, extent=[-122.4, -121.7, 37.5, 38.1])

# set aspect ratio to 1.26
ax.set_aspect(1.26)

# annotate facility numbering
for i in range(17):
    ax.annotate(i, (alameda_longs[i], alameda_lats[i]), fontsize=36)              
for i in range(8):
    ax.annotate(i+17, (contra_costa_longs[i], contra_costa_lats[i]), fontsize=36)
    
plt.savefig("visualizations/facility-scatter/facility_overlay.png")
plt.savefig("visualizations/facility-scatter/facility_overlay.pdf")
```

We also want a zoomed-in map for facilities near Oakland.

```python
oakland_fac_indices = [1, 2, 4, 11, 14]
```

```python
fig = plt.figure(figsize=(36, 20))
for i in oakland_fac_indices:
    plt.scatter(alameda_longs[i], alameda_lats[i], marker="o", color=alameda_colors[i], s=196)
ax = plt.axes()
plt.xlabel("Longitude", fontsize=36)
plt.ylabel("Latitude", fontsize=36)
ax.ticklabel_format(style='plain', useOffset=False)

plt.locator_params(axis="x", nbins=3)
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(36)
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(36)

# show map in background
oakland_img = plt.imread("all-data/gmaps-images/oakland_map_resized.png")
ax.imshow(oakland_img, alpha=.5, extent=[-122.275, -122.250, 37.81, 37.87])

# set aspect ratio to 1.26
ax.set_aspect(1.26)

# annotate facility numbering
for i in oakland_fac_indices:
    ax.annotate(i, (alameda_longs[i], alameda_lats[i]), fontsize=36)
    
plt.savefig("visualizations/facility-scatter/oakland_facility_overlay.png")
plt.savefig("visualizations/facility-scatter/oakland_facility_overlay.pdf")
```
